### Question 58
Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

Unknown : N

Less than $40K : a

$40K - $60K : b

$60K - $80K : c

$80K - $120K : d

$120K +’ : e

```python
def changeCategory(x):
    if x == "Unknown":
        return 'N'
    elif x == "Less than $40K":
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    elif x == '$120K+':
        return 'e'

df_card.Income_Category.apply(changeCategory).head(5)
```

### Question 67
Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값을 100을 빼서 새롭게 날짜를 Yr_Mo_Dy컬럼에 정의하라

```python
def newDatetime(x):
    import datetime
    if x.year >= 2061:
        new_year = x.year - 100
        return datetime.date(new_year, x.month, x.day)
    else:
        return x


df_stock['Yr_Mo_Dy'].apply(newDatetime)
```

```python
def fix_year(x):
    import datetime
    year = x.year - 100 if x.year >= 2061 else x.year
    return pd.to_datetime(datetime.date(year, x.month, x.day))

df_stock['Yr_Mo_Dy'] = df_stock['Yr_Mo_Dy'].apply(fix_year)
```
