### Question 23
quantity, item_price 두개의 컬럼으로 구성된 새로운 데이터 프레임을 정의하라.

```python
Ans = df_dining[['quantity', 'item_price']]
Ans.head()
```

### Question 24
item_price 컬럼의 달러표시 문자를 제거하고 flaot타입으로 저장하여 new_price 컬럼에 저장하라.

```python
## ver 1
Ans['new_price'] = Ans.item_price.str.replace('$', '').astype('float')
Ans.new_price.head(5)
```

```python
## ver 2
Ans['new_price'] = Ans.item_price.str[1:].astype('float')
Ans['new_price'].head(5)
```

### Question 25

```python
## ver1
Ans[Ans.new_price <= 5].shape[0]
```

```python
## ver2
len(Ans[Ans.new_price <=5])
```

### Question 26

### Question 27
new_price값이 9 이하이고 item_name 값이 Chicken Salad Bowl인 데이터 프레임을 추출하라

```python
Ans['item_name'] = df_dining['item_name']
Ans['new_price'] = df_dining['item_price'].str[1:].astype('float')
Ans[(Ans.item_name == 'Chicken Salad Bowl') & (Ans.new_price <= 9)].head(3)
```

### Question 28
