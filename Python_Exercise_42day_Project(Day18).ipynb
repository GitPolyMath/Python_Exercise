### Question 58
Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

Unknown : N

Less than $40K : a

$40K - $60K : b

$60K - $80K : c

$80K - $120K : d

$120K +’ : e

```python
def changeCategory(x):
    if x == "Unknown":
        return 'N'
    elif x == "Less than $40K":
        return 'a'
    elif x == '$40K - $60K':
        return 'b'
    elif x == '$60K - $80K':
        return 'c'
    elif x == '$80K - $120K':
        return 'd'
    elif x == '$120K+':
        return 'e'

df_card.Income_Category.apply(changeCategory).head(5)
```

### Question 67
Yr_Mo_Dy에 년도가 2061년 이상의 경우에는 모두 잘못된 데이터이다. 해당경우의 값을 100을 빼서 새롭게 날짜를 Yr_Mo_Dy컬럼에 정의하라

```python
def newDatetime(x):
    import datetime
    if x.year >= 2061:
        new_year = x.year - 100
        return datetime.date(new_year, x.month, x.day)
    else:
        return x


df_stock['Yr_Mo_Dy'].apply(newDatetime)
```

### Question 76
년-월-일:시 컬럼을 pandas에서 인식할 수 있는 datetime 형태로 변경하라. 서울시의 제공데이터의 경우 0시가 24시로 표현된다

```python
df_dust =pd.read_csv('https://raw.githubusercontent.com/Datamanim/pandas/main/seoul_pm.csv')
df_dust.head()
```

```python
df_dust.info()
```

```python
dt = []
for date, time in df_dust['(년-월-일:시)'].str.split(':'):
    if time == '24':
        time = '00'
    dt.append(date+' '+time+':00:00')

df_dust['년-월-일:시'] = pd.to_datetime(pd.Series(dt))
df_dust.head()
```

```python
def change_date(x):
    import datetime
    hour = x.split(':')[1]
    date = x.split(':')[0]

    if hour == '24':
        hour = '00:00:00'

        Datetime = pd.to_datetime(date+' '+hour)+datetime.timedelta(days=1)
    else:
        hour = hour+':00:00'
        Datetime = pd.to_datetime(date+' '+hour)
    return Datetime

df_dust['(년-월-일:시)'] = df_dust['(년-월-일:시)'].apply(change_date)
df_dust.head(5)
```
