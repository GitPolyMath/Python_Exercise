### Question 33
df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으며 제거하되 첫번째 케이스만 남겨라

```python
df_dining[(df_dining.item_name == 'Steak Salad') | (df_dining.item_name == 'Bowl')].drop_duplicates(subset = ['item_name'], keep='first')
```

### Question 34
df의 item_name 컬럼 값이 Steak Salad 또는 Bowl인 데이터를 데이터 프레임화 한 후, item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
## DataFrame.drop_duplicates(subset=None, *, keep='first', inplace=False, ignore_index=False)
## subset = 특정 열 기준
## keep = 유지할 중복값 선
## inplace = 데이터 프레임을 수정할지 여부
## ignore_index = index 초기화

df_dining[(df_dining.item_name == 'Steak Salad') | (df_dining.item_name == 'Bowl')].drop_duplicates(subset = 'item_name', keep='last')
```
