## 03. Grouping


### Question 44
데이터를 로드하고 상위 5개 컬럼을 출력하라.

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df_air = pd.read_csv(DataUrl)
df_air.head(5)
```

### Question 45
데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
df_air.host_name.value_counts().sort_index(ascending = True).head(5)
```

```python
df_air.groupby('host_name').size().head()
```

### Question 46
데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라

```python
df_air.host_name.value_counts().sort_values(ascending=False).to_frame().rename(columns={'host_name':'counts'}).head(5)
```

```python
df_air.groupby('host_name').size().sort_values(ascending=False).to_frame().rename(columns={0:'counts'}).head(5)
```

### Question 47
neighbourhood_group의 값에 따른 neighbourhood 컬럼 값의 갯수를 구하여라.

```python
df_air.groupby(['neighbourhood_group', 'neighbourhood']).size().to_frame().rename(columns={0:'size'}).head()
```

```python
## groupby(by=[], as_index = ) as_index : 집계된 출력의 경우 그룹 레이블이 있는 개체를 인덱스로 반환합니다.
## 즉. "SQL"스타일의 groupby
df_air.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().head()
```

### Question 48
neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최대값들을 출력하라.

```python
df_air.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby('neighbourhood_group', as_index=False).agg('max')
```

### Question 49
neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라

```python
df_air[['neighbourhood_group','price']].groupby('neighbourhood_group').agg(['mean','var','max','min'])
```

### Question 50
neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라

```python
df_air[['neighbourhood_group','reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
```
